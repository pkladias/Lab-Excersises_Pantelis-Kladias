## Part 1

```python
from tkinter import  *
import requests 
import json
```

```python
authors_request = requests.get(f'https://api.quotable.io/authors?sortBy=name')
read_content = authors_request.json()
read_content
```

```python
author_access = read_content['results']
author_access
```

```python
author_list = []
for author_data in author_access:
    author_list.append(author_data['name'])
```

```python
win= Tk()


win.geometry("750x250")

def get_quote():
    quote = requests.get(f'https://api.quotable.io/random/?author=Abraham Lincoln')
    jsn = quote.json()
    jsn
    quote_content = jsn['content']
    x = quote_content
    return x

quote= get_quote()
# I could not get it to generate a quote from a specific person so I just used Lincoln

def my_click():
    mylabel=Label(win, text= quote)
    mylabel.pack()

menu= StringVar()
menu.set("Select Any Person You would Like to see a Quote From")

drop= OptionMenu(win, menu, *author_list)
drop.pack()

my_button = Button(win, text= "Get Quote", command = my_click).pack()

win.mainloop()
```

```python
import requests 
import json
import pandas as pd
import csv
```

```python
lat = "42.1670"
lon = "-76.8205"

genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")

json_file = genius.json()
json_file
```

```python
forecast = json_file["properties"]["forecast"]
new_request = requests.get(forecast)

    
json = new_request.json()
json
```

```python
a =json["properties"]["periods"]
a
```

```python
days = []
temp = []
detail = []
for i in a:
    days.append(i['name'])
    temp.append(i["temperature"])
    detail.append(i["detailedForecast"])
```
